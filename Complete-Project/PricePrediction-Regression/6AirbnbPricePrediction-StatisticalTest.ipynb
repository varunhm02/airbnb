# Statistical Test - Paired t-test

```python
from scipy.stats import ttest_ind
import pandas as pd

df = pd.read_csv('price_prediction_results_80.csv')
df
```

```python

col = df.columns
results_frame = pd.DataFrame()
row = []
for i in range(len(df.columns)):
   
    for j in range(i+1, len(df.columns)):
        column = str(col[i]) +' VS '+ str(col[j]) 
        t_val, p_val = ttest_ind(df[df.columns[i]],df[df.columns[j]])
        results_frame[column] = [p_val]
        if p_val < 0.05 :
            row.append('S')
        else:
            row.append('-')

results_frame.loc[len(results_frame)] = row
        
results_frame.head()
```

```python
results_frame.to_csv('statresults_80.csv',index=False)
```
