We see here that data types for price and cleaning fee and security_deposit are not right, they are supposed to be numbers but they are object here.



```python
data['price'] = data['price'].str.replace(',', '')
data['price'] = data['price'].str.replace('$', '')
data['price'] = data['price'].astype(float)


data['security_deposit'] = data['security_deposit'].str.replace(',', '')
data['security_deposit'] = data['security_deposit'].str.replace('$', '')
data['security_deposit'] = data['security_deposit'].astype(float)

data['cleaning_fee'] = data['cleaning_fee'].str.replace(',', '')
data['cleaning_fee'] = data['cleaning_fee'].str.replace('$', '')
data['cleaning_fee'] = data['cleaning_fee'].astype(float)

```

Checking which features have null values and how many?

```python
data.isnull().sum()
```

Many columns like thumbnail_url, medium_url, xl_picture_url, host_acceptance_rate, neighbourhood_group_cleansed, license have all null values. 

Moreover columns like  weekly_price, monthly_price, and square_feet are almost filled with null values

We can drop the above mentioned columns

There are many other columns that have significant number of null values such as security_deposit, cleaning_fee, host_about, host_response_time, host_response_rate, space,neighborhood_overview, transit,access, notes, interaction, house_rules. Such columns need further inspection whether they are important or not , whether they can be imputed or not, can be decided later


```python
drop_columns=['thumbnail_url', 'medium_url', 'xl_picture_url', 'host_acceptance_rate', 'neighbourhood_group_cleansed', 'license', 'square_feet','weekly_price','monthly_price']

data = data.drop(drop_columns,axis=1)
```

Lets dig deeper and look into the values the important features hold

Starting with the country feature

```python
data.country.value_counts()
```

```python
data[data['country']=='France']
```

We can see there are 3 instances that have their country set as France, if we further look into the data for those rows, we find that the listing belongs to Canada only. Since there were just 3 such case we could manually look up and verify, if there were more such noisy data, best option would be to drop such columns as we have enough data.

```python
# if you want to delete rows with France as country
#data = data[data.country == 'Canada']

#change France to Canada
data.country.replace('France','Canada',inplace=True)

#verifying the modification
data.country.value_counts()
```
